```python
data['date']=pd.to_datetime(data['date'])
print(data.dtypes)
data.head(5)
```

```python
data=data.drop(data.columns[5:], axis=1)
data.head(5)
```

```python
from sklearn.cluster import KMeans
import numpy as np
 
route_stats = data.groupby('route_id')['count'].mean().reset_index()
 
kmeans = KMeans(n_clusters=3, random_state=42)
route_stats['cluster'] = kmeans.fit_predict(route_stats[['count']])
 
data = data.merge(route_stats[['route_id', 'cluster']], on='route_id', how='left')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
scaler = StandardScaler()
day_data2=data[data['hour']<20]
day_data2=data[data['hour']>6]
training_data=day_data2[['count','hour','dayOfWeek', 'cluster']]
training_data['count'] = scaler.fit_transform(training_data[['count']])

training_data.dtypes
clf = IsolationForest(contamination=0.01, random_state=42)
clf.fit(training_data)
y_pred = clf.predict(training_data)

```
